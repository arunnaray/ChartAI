```python
import pandas as pd
import numpy as np
```

```python
required_columns = ['work_location', 'work_country', 'business_unit', 'department', 'designation', 'band']
```

```python
work_location = ['Building 1', 'Building 2', 'Building 3', 'Building 4', 'Building 5',
                'Building 6', 'Building 7', 'Building 8', 'Building 9', 'Building 10']
```

```python
work_country = ['India', 'USA', 'UK', 'France', 'Germany', 'China', 'Japan', 'Australia',
                'Canada', 'Brazil', 'Mexico', 'Argentina', 'South Africa', 'Nigeria', 'Kenya']
```

```python
business_unit = ['BU1', 'BU2', 'BU3', 'BU4', 'BU5', 'BU6']
```
